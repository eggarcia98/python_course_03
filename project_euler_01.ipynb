1. If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
limit_number = 1000

valid_numbers = []
for n in range(1, limit_number):
  is_multiple_of_3 = n % 3 == 0
  is_multiple_of_5 = n % 5 == 0

  if is_multiple_of_3 or is_multiple_of_5:
    valid_numbers.append(n)

sum_of_valid_numbers = sum(valid_numbers)

print(f"The sum of all the multiples of 3 or 5 bellow {limit_number}: {sum_of_valid_numbers}")
```

3. The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?


```python
def get_prime_factors(n):
    factors = []
    i = 2
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

number = 600851475143
prime_factors = get_prime_factors(number)
print("The Largest Prime factors of", number, "is:", max(prime_factors))
```

Find the largest palindrome made from the product of two 3-digit numbers.


```python


def is_palindrome_number(number):
  string_number = str(number)
  reversed_number =  string_number[::-1]

  if(string_number == reversed_number):
    return True
  else: 
    return False
```

```python
import itertools

numbers_with_3_digits = list(range(100, 999 + 1))
combinations = list(itertools.combinations(numbers_with_3_digits, 2))

products = [list(x)[0] * list(x)[1] for x in combinations]

palindrome_products = filter(is_palindrome_number, products)
largest_palindrome = max(palindrome_products)

print(f"The Largest Palindrome made from the product of two 3-digit numbers: {largest_palindrome}")
```

5. 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?


```python
#Return dict when key is prime factor and value its potency
def get_prime_factors(number):

  divisors_dict = dict()

  current_divisor = 2
  current_divident = number

  while current_divident != 1:

    if current_divident % current_divisor == 0:
      if(divisors_dict.get(current_divisor) == None):
        divisors_dict[current_divisor] = 0
      
      current_divident = current_divident / current_divisor
      divisors_dict[current_divisor] += 1
    else:
      current_divisor += 1

  return divisors_dict
```

```python
from functools import reduce

divisibles_numbers_to_check = list(range(1,21))

factors_summary = dict()
for divisible_number in divisibles_numbers_to_check:
  factors = get_prime_factors(divisible_number)

  for factor, potency in factors.items():
    if potency > int(factors_summary.get(factor) or 0):
        factors_summary[factor] = potency

smallest_number = reduce(
      lambda acc, item: acc * pow(list(item)[0], list(item)[1]), 
      factors_summary.items(), 
      1
    )

print(f"Smallest Number: {smallest_number}" )
```

6. The sum of the squares of the first ten natural numbers is, 1^2 + 2^2 + ... + 10^2 = 385

The square of the sum of the first ten natural numbers is, (1 + 2 + ... + 10)^2 = 3025

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is, 3025 - 385 = 2640

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
from functools import reduce

max_number = 100
numbers = list(range(1, max_number + 1))

sum_of_squares = reduce(lambda x, number: x + number**2, numbers, 0)
square_of_sum = reduce(lambda x, number: x + number, numbers, 0)**2

diff = square_of_sum - sum_of_squares
print(f"The difference between the sum of the squares of the first {max_number} natural numbers and the square of the sum is {diff}")
```

7. By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10001st prime number?


```python
import math

def is_prime(n):
  if n == 2 or n == 3:
      return True
  if n < 2 or n % 2 == 0:
      return False
  
  # Using the integer square root for efficiency
  sqrt_n = math.isqrt(n)  

  # Only evalue odd's divisor becouse even divisor if divisible to 2...
  # ... and alredy this posibility is controled with second if of this function
  for divisor in range(3, sqrt_n + 1, 2):
      if n % divisor == 0:
          return False

  return True

objetive_number_position = 1001 # You can change this parameter
number_to_evaluate = 1

prime_numbers = []
while len(prime_numbers) != objetive_number_position:
  if is_prime(number_to_evaluate):
    prime_numbers.append(number_to_evaluate)
  
  number_to_evaluate += 1

print(f" the {objetive_number_position}st prime number: {prime_numbers[-1]}")

```

8. A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
a^2 + b^2 = c^2

For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.


```python
# Equation 
# c = 1000 - a - b

a,b,c = [0,0,0]

# a can't be grater than b and c, so can't be greater than 350
for a_value in range(1, 333):
    for b_value in range(a_value + 1, 500):
        (a, b) = [a_value, b_value]
        
        c = 1000 - a - b
        if a ** 2 + b ** 2 == c ** 2:
            break
    
    if a ** 2 + b ** 2 == c ** 2:
      break

print(f"Product of a: {a}, b: {b}, c: {c}:", a*b*c)


```

9. The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.


```python
# the Sieve of Eratosthenes algorithm
def sum_of_primes(limit):
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False
    p = 2
    while p * p <= limit:
        if sieve[p]:
            for i in range(p * p, limit + 1, p):
                sieve[i] = False
        p += 1

    sum_of_primes = sum(i for i in range(limit + 1) if sieve[i])
    return sum_of_primes

limit = 10
result = sum_of_primes(limit)
print("Sum of primes below two million:", result)
```
